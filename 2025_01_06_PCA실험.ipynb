### **수집한 데이터로 행동을 판별할 수 있는가?**

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# CSV 파일 로드
run_data = pd.read_csv('RUN_total.csv')
walk_data = pd.read_csv('WALK_total.csv')
left_data = pd.read_csv('LEFT_total.csv')
right_data = pd.read_csv('RIGHT_total.csv')

# timestamp 열 삭제
run_data = run_data.drop(columns=['timestamp'])
walk_data = walk_data.drop(columns=['timestamp'])
left_data = left_data.drop(columns=['timestamp'])
right_data = right_data.drop(columns=['timestamp'])

mag_features = ['X_ORI', 'Y_ORI', 'Z_ORI']
run_data = run_data.drop(columns=mag_features)
walk_data = walk_data.drop(columns=mag_features)
left_data = left_data.drop(columns=mag_features)
right_data = right_data.drop(columns=mag_features)

# 레이블 추가
run_data['Label'] = 'Running'
walk_data['Label'] = 'Walking'
left_data['Label'] = 'Left'
right_data['Label'] = 'Right'

# 데이터 병합
data = pd.concat([run_data, walk_data, left_data, right_data], ignore_index=True)

# 피처와 레이블 분리
X = data.drop(columns=['Label'])  # 피처만 남김
y = data['Label']  # 레이블

# 데이터 스케일링
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA 수행 (2차원)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# PCA 결과를 DataFrame으로 변환
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['Label'] = y

# PCA 시각화
plt.figure(figsize=(8, 6))
for label, color in zip(['Running', 'Walking', 'Left', 'Right'], ['red', 'blue', 'yellow', 'black']):
    subset = pca_df[pca_df['Label'] == label]
    plt.scatter(subset['PC1'], subset['PC2'], label=label, alpha=0.6, edgecolors='k')

plt.title('PCA of IMU Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

# **LPF로 노이즈 줄이고 PCA**

```python
# LPF Algorithm
from scipy.signal import butter, filtfilt

# Low Pass Filter 함수 정의
def low_pass_filter(data, cutoff=0.1, fs=1.0, order=2):
    """
    data: 필터를 적용할 데이터 (numpy 배열)
    cutoff: 차단 주파수 (기본값 0.1)
    fs: 샘플링 주파수 (기본값 1.0)
    order: 필터 차수 (기본값 2)
    """
    nyquist = 0.5 * fs  # 나이퀴스트 주파수
    normal_cutoff = cutoff / nyquist  # 정규화된 차단 주파수
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return filtfilt(b, a, data)
```

## **PCA with LPF + RAW**

```python
# CSV 파일 로드
run_data = pd.read_csv('RUN_total.csv')
walk_data = pd.read_csv('WALK_total.csv')
left_data = pd.read_csv('LEFT_total.csv')
right_data = pd.read_csv('RIGHT_total.csv')

# timestamp 열 삭제
run_data = run_data.drop(columns=['timestamp'])
walk_data = walk_data.drop(columns=['timestamp'])
left_data = left_data.drop(columns=['timestamp'])
right_data = right_data.drop(columns=['timestamp'])


CUTOFF = 0.2
FS = 100
# Low Pass Filter 적용 (모든 컬럼에 대해 필터링)
for col in run_data.columns:
    run_data[col] = low_pass_filter(run_data[col].values, cutoff=CUTOFF, fs=FS, order=2)
    walk_data[col] = low_pass_filter(walk_data[col].values, cutoff=CUTOFF, fs=FS, order=2)
    left_data[col] = low_pass_filter(left_data[col].values, cutoff=CUTOFF, fs=FS, order=2)
    right_data[col] = low_pass_filter(right_data[col].values, cutoff=CUTOFF, fs=FS, order=2)

# 레이블 추가
run_data['Label'] = 'Running'
walk_data['Label'] = 'Walking'
left_data['Label'] = 'Left'
right_data['Label'] = 'Right'

# 데이터 병합
data = pd.concat([run_data, walk_data, left_data, right_data], ignore_index=True)

# 피처와 레이블 분리
X = data.drop(columns=['Label'])  # 피처만 남김
y = data['Label']  # 레이블

# 데이터 스케일링
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA 수행 (2차원)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# PCA 결과를 DataFrame으로 변환
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['Label'] = y

# PCA 시각화
plt.figure(figsize=(8, 6))
for label, color in zip(['Running', 'Walking', 'Left', 'Right'], ['red', 'blue', 'yellow', 'black']):
    subset = pca_df[pca_df['Label'] == label]
    plt.scatter(subset['PC1'], subset['PC2'], label=label, alpha=0.05, edgecolors='k', s=100)

plt.title('PCA of IMU Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

* **LPF를 적용하니 좌, 우 회전이 구분되기 시작했다**
